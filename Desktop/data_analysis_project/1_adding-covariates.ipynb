```python
#IMPUTATION: Mean imputation by group
# Columns to convert
cols_to_imp = ['medianIncome','inLaborForce', 'employed', 'unemployed',
       'privateHI<19', 'publicHI<19', 'privatePublicHI<19', 'noHI<19',
       'privateHI19-34', 'publicHI19-34', 'privatePublicHI19-34', 'noHI19-34',
       'privateHI35-64', 'publicHI35-64', 'privatePublicHI35-64', 'noHI35-64',
       'privateHI65+', 'publicHI65+', 'privatePublicHI65+', 'noH65+']

df_merged3[cols_to_imp] = df_merged3.groupby('STATE')[cols_to_imp].transform(lambda x: x.fillna(x.mean()))
```

```python
import numpy as np
#preprocess data
#1) sum health insurance var
df_merged3['privateSum'] = df_merged3[['privateHI<19', 'privateHI19-34', 'privateHI35-64','privateHI65+']].sum(axis=1)
df_merged3['publicSum'] = df_merged3[['publicHI<19', 'publicHI19-34', 'publicHI35-64','publicHI65+']].sum(axis=1)
df_merged3['privatepublicSum'] = df_merged3[['privatePublicHI<19', 'privatePublicHI19-34', 'privatePublicHI35-64','privatePublicHI65+']].sum(axis=1)
df_merged3['noHISum'] = df_merged3[['noHI<19', 'noHI19-34', 'noHI35-64','noH65+']].sum(axis=1)

#2) make percent vars
origvar=["white","black","AIAN","asian","NHPI","otherRace","twoplusRace","notHispanicLatino","hispanicLatino","employed","unemployed","privateSum","publicSum","privatepublicSum","noHISum"]
newvar=["white%","black%","AIAN%","asian%","NHPI%","otherRace%","twoplusRace%","notHispanicLatino%","hispanicLatino%","employed%","unemployed%","private%","public%","privatepublic%","noHI%"]

for index, element in enumerate(origvar):
    df_merged3[newvar[index]] = df_merged3[element] / df_merged3['totalPop']

#3) make pre/post covid binary var
df_merged3['covidBin'] = np.where(df_merged3['YEAR'] > 2019, 1, 0)
```

```python
#EXPORT (post-imputation)
df_merged3.to_excel(cd+'enrollment_expenditure_covar_imp_df.xlsx', index=False)
```
